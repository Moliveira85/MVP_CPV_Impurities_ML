## 1.1 Descrição do Problema

<br>

A indústria farmacêutica possui a responsabilidade de garantir que seus processos produtivos permaneçam estáveis e capazes de gerar produtos com a qualidade esperada ao longo de todo o seu ciclo de vida. Para isso, são realizadas atividades de Continued Process Verification (CPV), que envolvem o monitoramento contínuo de atributos de qualidade e parâmetros de processo.

Com o avanço da digitalização na indústria farmacêutica e o aumento do volume de dados gerados por laboratórios e áreas produtivas, surge a oportunidade de utilizar técnicas de Inteligência Artificial (IA) e Machine Learning para apoiar a análise desses dados e identificar padrões que possam auxiliar atividades de monitoramento e tomada de decisão.

Neste trabalho, o modelo desenvolvido tem como objetivo realizar uma **previsão**. Mais especificamente, busca-se prever um atributo de qualidade do produto utilizando dados históricos de matéria-prima e processo. Como estudo de caso, foi selecionada a variável **Impurezas Totais (impurities_total)**, utilizada como exemplo para avaliar a capacidade preditiva dos modelos desenvolvidos.

Os principais interessados nessa solução seriam profissionais das áreas de Qualidade, Controle de Qualidade (CQ), Process Validation, Desenvolvimento Farmacêutico e equipes responsáveis pelas atividades de CPV. Esses profissionais poderiam utilizar modelos preditivos como ferramenta complementar para monitoramento de tendências e avaliação de risco.

A relevância deste problema está no potencial de utilização da Inteligência Artificial como suporte às atividades de Continued Process Verification, permitindo análises mais rápidas e orientadas por dados. Embora o presente estudo utilize Impurezas Totais como atributo de qualidade, a metodologia proposta pode ser aplicada a outros Critical Quality Attributes (CQAs), como teor, dissolução, umidade, uniformidade de conteúdo ou impurezas específicas, desde que dados históricos adequados estejam disponíveis.

Dessa forma, a pergunta central deste trabalho é:

## 1.3 Tipo de Problema


**Tipo escolhido:** Regressão

**Justificativa:**

O presente estudo foi definido como um problema de regressão, pois o objetivo do modelo é prever um valor numérico contínuo. A variável alvo utilizada foi **Impurezas Totais (`impurities_total`)**, representada por valores quantitativos que podem assumir diferentes magnitudes dentro de uma faixa contínua.

Diferentemente de um problema de classificação, em que o objetivo seria prever categorias como "Aprovado/Reprovado" ou "Alto/Baixo Risco", neste trabalho busca-se estimar diretamente o valor esperado das impurezas totais a partir de atributos de matéria-prima e processo.

Também não se trata de um problema de clusterização, uma vez que o conjunto de dados possui uma variável alvo conhecida utilizada para o treinamento supervisionado dos modelos.

Por fim, o estudo não foi tratado como um problema de séries temporais, pois o objetivo não é prever valores futuros respeitando uma sequência cronológica específica, mas sim identificar relações entre variáveis de entrada e o atributo de qualidade analisado.

Assim, a abordagem de regressão foi considerada a mais adequada para avaliar o potencial de técnicas de Machine Learning como suporte às atividades de Continued Process Verification (CPV), utilizando as Impurezas Totais como estudo de caso.

```python
# Importação das métricas utilizadas para avaliação dos modelos
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# Biblioteca utilizada para cálculos matemáticos
import numpy as np

# Função auxiliar para cálculo das métricas de desempenho
def avaliar_modelo(y_true, y_pred):

    mae = mean_absolute_error(y_true, y_pred)

    rmse = np.sqrt(
        mean_squared_error(y_true, y_pred)
    )

    r2 = r2_score(y_true, y_pred)

    return mae, rmse, r2
```

```python
# Consolidação das métricas de desempenho dos modelos avaliados
resultados = pd.DataFrame({
    "Modelo": [
        "Baseline",
        "Linear Regression",
        "Random Forest",
        "Gradient Boosting"
    ],
    "MAE": [
        avaliar_modelo(y_test, pred_base)[0],
        avaliar_modelo(y_test, pred_lr)[0],
        avaliar_modelo(y_test, pred_rf)[0],
        avaliar_modelo(y_test, pred_gb)[0]
    ],
    "RMSE": [
        avaliar_modelo(y_test, pred_base)[1],
        avaliar_modelo(y_test, pred_lr)[1],
        avaliar_modelo(y_test, pred_rf)[1],
        avaliar_modelo(y_test, pred_gb)[1]
    ],
    "R²": [
        avaliar_modelo(y_test, pred_base)[2],
        avaliar_modelo(y_test, pred_lr)[2],
        avaliar_modelo(y_test, pred_rf)[2],
        avaliar_modelo(y_test, pred_gb)[2]
    ]
})

# Ordenação dos modelos pelo melhor coeficiente de determinação (R²)
resultados.sort_values(
    by='R²',
    ascending=False
)
```

# 9. Validação e Otimização de Hiperparâmetros

## 9.1 Modelo Selecionado para Otimização

O modelo escolhido para otimização foi o Random Forest Regressor.

A escolha foi motivada pelo desempenho superior apresentado durante as etapas anteriores do estudo, além de sua capacidade de capturar relações não lineares entre os atributos disponíveis e o atributo de qualidade analisado.

## 9.2 Estratégia de Busca

Foi utilizada a técnica Grid Search para avaliar diferentes combinações de hiperparâmetros do algoritmo Random Forest.

A busca foi mantida propositalmente simples e compatível com os objetivos do MVP, priorizando a demonstração da metodologia de otimização sem gerar custo computacional excessivo.

```python
# Definição dos hiperparâmetros avaliados durante a otimização do modelo
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5]
}

# Configuração do Grid Search com validação cruzada
grid_rf = GridSearchCV(
    estimator=RandomForestRegressor(
        random_state=42
    ),
    param_grid=param_grid,
    scoring='r2',
    cv=5,
    n_jobs=-1
)

# Execução da busca pelos melhores hiperparâmetros
grid_rf.fit(
    X_train_v3,
    y_train_v3
)
```

```python
# Exibição da melhor combinação de hiperparâmetros encontrada
print("Melhores parâmetros:")
print(grid_rf.best_params_)

# Exibição do melhor resultado obtido na validação cruzada
print("\nMelhor R² CV:")
print(grid_rf.best_score_)
```

```python
# Seleção do modelo com melhor desempenho encontrado pelo Grid Search
best_rf = grid_rf.best_estimator_
```

```python
# Geração das previsões do modelo otimizado nos conjuntos de treino e teste
pred_train_opt = best_rf.predict(X_train_v3)
pred_test_opt = best_rf.predict(X_test_v3)
```

```python
# Importação das métricas utilizadas na avaliação final do modelo
from sklearn.metrics import (
    r2_score,
    mean_absolute_error,
    mean_squared_error
)

# Biblioteca utilizada para cálculos matemáticos
import numpy as np

# Avaliação do desempenho do modelo otimizado
print("R² Treino:", r2_score(y_train_v3, pred_train_opt))
print("R² Teste :", r2_score(y_test_v3, pred_test_opt))

print("MAE:", mean_absolute_error(y_test_v3, pred_test_opt))

print("RMSE:", np.sqrt(mean_squared_error(y_test_v3, pred_test_opt)))
```
